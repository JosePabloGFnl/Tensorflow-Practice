```python
%tensorflow_version 2.x #always start with this when you want to use tensorflow
import tensorflow as tf
print(tf.version)
```

```python
#creating tensors

#rank 0 tensors
string = tf.Variable("this is a string", tf.string)
number = tf.Variable(324, tf.int16)
floating = tf.Variable(3.567, tf.float64)

#ranks/degrees of tensors
rank1_tensor = tf.Variable(["Test","Hello","Pablo"], tf.string) #one list, one array, one dimension
rank2_tensor = tf.Variable([["Test","Hello"],["Test","Rank 2"]], tf.string)
```

```python
#to determine the rank of a tensor
tf.rank(number)
```

```python
tf.rank(rank1_tensor)
```

```python
tf.rank(rank2_tensor)
```

```python
#changing shapes
tensor1 = tf.ones([1,2,3]) #tf.ones() creates a shape [1,2,3] tensor full of ones
tensor2 = tf.reshape(tensor1, [2,3,1]) #reshape existing data to shape [2,3,1]
tensor3 = tf.reshape(tensor2, [3, -1]) #-1 tells the tensor to calculare the size of the dimension in that place
                                       # this will reshape the tensor to [3,2]
#the number of elements in the reshape tensor must match the number in the original
```

```python
print("Tensor 1:")
print(tensor1)
print("Tensor 2:")
print(tensor2)
print("Tensor 3:")
print(tensor3)
```
